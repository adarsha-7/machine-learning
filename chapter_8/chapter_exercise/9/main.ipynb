# Random Forest 

```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1)

X = mnist.data
y = mnist.target

X_train, X_test = X[:60000], X[60000:]
y_train, y_test = y[:60000], y[60000:]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
import time

rf = RandomForestClassifier(random_state=42)
rf_start = time.time()
rf.fit(X_train, y_train)
rf_end = time.time()

rf_pred = rf.predict(X_test)

print("Training Time: ", rf_end - rf_start)
print("Accuracy: ", accuracy_score(rf_pred, y_test))
```

```python
pca = PCA(n_components=0.95)
X_train_red = pca.fit_transform(X_train)
X_test_red = pca.transform(X_test)
pca.n_components_
```

```python
pca_rf = RandomForestClassifier(random_state=42)
pca_rf_start = time.time()
pca_rf.fit(X_train_red, y_train)
pca_rf_end = time.time()

pca_rf_pred = pca_rf.predict(X_test_red)

print("Training Time: ", pca_rf_end - pca_rf_start)
print("MSE: ", accuracy_score(pca_rf_pred, y_test))
```
