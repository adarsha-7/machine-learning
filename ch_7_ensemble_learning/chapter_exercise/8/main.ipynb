```python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1)

X = mnist.data
y = mnist.target

len(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    train_size=50000,
    random_state=42
)

X_test, X_valid, y_test, y_valid = train_test_split(
    X_test, y_test,
    train_size=10000,
    random_state=42
)

print(len(X_train))
print(len(X_test))
print(len(X_valid))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.svm import LinearSVC

rnd_clf = RandomForestClassifier(random_state=42, n_estimators=100, n_jobs=-1)
etrees_clf = ExtraTreesClassifier(random_state=42, n_estimators=100, n_jobs=-1)
svc_clf = LinearSVC(random_state=42, max_iter=100, dual=False)

rnd_clf.fit(X_train, y_train)
etrees_clf.fit(X_train, y_train)
svc_clf.fit(X_train, y_train)
```

```python
print(rnd_clf.score(X_valid, y_valid))
print(etrees_clf.score(X_valid, y_valid))
print(svc_clf.score(X_valid, y_valid))
```

```python
from sklearn.ensemble import VotingClassifier

voting_clf_hard = VotingClassifier(
    estimators=[
        ('rf', rnd_clf),
        ('et', etrees_clf),
        ('svc', svc_clf),
    ],
    voting='hard',
    n_jobs=-1
)

voting_clf_hard.fit(X_train, y_train)
```

```python
voting_clf_hard.score(X_valid, y_valid)
```

```python
rnd_clf = RandomForestClassifier(random_state=42, n_estimators=100, n_jobs=-1)
etrees_clf = ExtraTreesClassifier(random_state=42, n_estimators=100, n_jobs=-1)

rnd_clf.fit(X_train, y_train)
etrees_clf.fit(X_train, y_train)
```

```python
voting_clf_hard = VotingClassifier(
    estimators=[
        ('rf', rnd_clf),
        ('et', etrees_clf),
    ],
    voting='hard',
    n_jobs=-1
)

voting_clf_hard.fit(X_train, y_train)

voting_clf_hard.score(X_valid, y_valid)
```

```python
voting_clf_soft = VotingClassifier(
    estimators=[
        ('rf', rnd_clf),
        ('et', etrees_clf),
    ],
    voting='soft',
    n_jobs=-1
)

voting_clf_soft.fit(X_train, y_train)

voting_clf_soft.score(X_valid, y_valid)
```

```python
voting_clf_soft.score(X_test, y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_clf = GradientBoostingClassifier(
    n_estimators=100, 
    random_state=42, 
    n_iter_no_change=10, 
    validation_fraction=0.1, 
    learning_rate=0.3,
    max_features='sqrt', 
    verbose=True)
gb_clf.fit(X_train, y_train)

gb_clf.score(X_valid, y_valid)
```

```python
from xgboost import XGBClassifier

xgb_clf = XGBClassifier(   
    objective='binary:logistic', 
    n_estimators=100,
    eval_metric="aucpr",
    early_stopping_rounds=10,
    random_state=42,
    device='cuda:0'
)

xgb_clf.fit(X_train, y_train.astype(int), 
            verbose=True, eval_set=[(X_valid, y_valid.astype(int))])

xgb_clf.score(X_valid, y_valid.astype(int))
```

```python
print(voting_clf_soft.score(X_test, y_test))
print(gb_clf.score(X_test, y_test))
print(xgb_clf.score(X_test, y_test.astype(int)))
```
