```python
import matplotlib.pyplot as plt
import numpy as np

# Years (x-axis)
years = np.arange(0, 11)  # 0 to 10

# Costs and revenue
fixed_cost = 80 * np.ones_like(years)
total_cost = 80 + 167.2 * years
revenue = 187.6 * years

# Scale values to 10^4
fixed_cost_scaled = fixed_cost * 1e3
total_cost_scaled = total_cost * 1e3
revenue_scaled = revenue * 1e3

# Correct Break-even point
break_even_x = 80 / (187.6 - 167.2)
break_even_y = 187.6 * break_even_x
break_even_y_scaled = break_even_y * 1e3

# Plot
plt.figure(figsize=(12, 8))

plt.plot(years, fixed_cost_scaled, marker='o', linewidth=2, label='Fixed Cost')
plt.plot(years, total_cost_scaled, marker='o', linewidth=2, label='Total Cost')
plt.plot(years, revenue_scaled, marker='o', linewidth=2, label='Revenue')

plt.scatter(break_even_x, break_even_y_scaled, color='red', s=200, zorder=5, label='Break-even')

plt.xlabel('Years')
plt.ylabel('NPR (in crores)')
plt.title('Cost, Revenue, and Break-even Analysis')

plt.xticks(years)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Percentage changes
percentages = np.array([-30, -20, -10, 0, 10, 20, 30])

# Scale factor
scale = 1e5  # values divided by 1000

# Data (divided by 1000 for scaling)
capital_investment = np.array([108591, 95007, 81423, 67839, 54254, 40670, 27086]) / scale
annual_revenue = np.array([-494961, -307361, -119761, 67839, 255438, 443038, 630638]) / scale
annual_expenses = np.array([569534, 402302, 235070, 67839, -99393, -266625, -433857]) / scale
useful_life = np.array([33907, 48223, 59194, 67839, 74783, 80458, 85159]) / scale
marr = np.array([182642, 157760, 119214, 67839, 6347, -62182, -135231]) / scale

# Create large figure
plt.figure(figsize=(16, 12))

# Plot lines with markers
plt.plot(percentages, capital_investment, marker='o', linewidth=2.5, label='Capital Investment')
plt.plot(percentages, annual_revenue, marker='o', linewidth=2.5, label='Annual Revenue')
plt.plot(percentages, annual_expenses, marker='o', linewidth=2.5, label='Annual Expenses')
plt.plot(percentages, useful_life, marker='o', linewidth=2.5, label='Useful Life')
plt.plot(percentages, marr, marker='o', linewidth=2.5, label='MARR')

# Axis formatting
plt.axvline(0)  # vertical line at base value
plt.xticks(percentages)
plt.grid(True, linestyle='--', alpha=0.6)

# Labels and title
plt.title('Sensitivity Analysis', fontsize=20)
plt.xlabel('Percentage Changed from Base Value (%)', fontsize=16)
plt.ylabel('Annual Worth (NPR in Lakhs)', fontsize=16)

plt.legend(fontsize=13)
plt.tight_layout()
plt.show()
```
